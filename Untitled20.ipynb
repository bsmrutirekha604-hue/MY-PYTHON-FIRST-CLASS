## 2.WAPP that accepts a string and prints the number of vowels in it. 

```python
text = input("Enter a string")
vowels = 'aeiou'
count = sum(1 for char in text if char.lower() in vowels)
print("no of vowels:", count)
```

## WAP that checks if the input number is a prime number.

```python
num = int(input("Enter a number:"))
if num > 1:
    for i in range(2,num):
       if num % i == 0:
           print("Not a prime number")
           break
       else:
            print("Prime number")
    else:
            print("Not a prime number") 
```

## WAP to calculate the factorial of a number using input().

```python
num = int(input("Enter a number:"))
factorial = 1 
for i in range(1,num+1):
    factorial *= i
    print("Factorial:",factorial)    
```
